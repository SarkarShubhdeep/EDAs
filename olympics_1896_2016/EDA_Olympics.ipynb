### 10. Find the Ratio of male and female athletes participated in all olympic games.

```python
df_male_female = pd.pivot_table(df.groupby(['Games', 'Sex']).count(), index='Games', columns='Sex')['Name']
df_male_female.fillna(0, inplace=True)
female_percent = round((df_male_female.F/(df_male_female.M + df_male_female.F)) * 100, 2)
male_percent = round((df_male_female.M/(df_male_female.M + df_male_female.F)) * 100, 2)
```

```python
df_male_female['Female %'] = female_percent
df_male_female['Male %'] = male_percent
```

### 17. Identify which country won the most gold, most silver, most bronze medals and the most medals in each olympic games.

```python
df_region = pd.pivot_table(df, index='Games', columns='Medal', aggfunc='max')['region'][['Gold', 'Silver', 'Bronze']]

df_medal_count = pd.pivot_table(df, index='Games', columns='Medal', aggfunc='count')['region'][['Gold', 'Silver', 'Bronze']]

gold = df_region.Gold + ', ' +list(map(str, (df_medal_count.Gold)))
silver = df_region.Silver + ', ' +list(map(str, (df_medal_count.Silver)))
bronze = df_region.Bronze + ', ' +list(map(str, (df_medal_count.Bronze)))

df_most_medal_country = pd.DataFrame(index=list(df_region.index))

```

```python
df_most_medal_country['Most_gold'] = list(gold)
df_most_medal_country['Most_silver'] = list(silver)
df_most_medal_country['Most_bronze'] = list(bronze)
```

## Conclusions/Insights

1. From 1896 to 2016, there are 51 Olympic games happened.
2. In 2016, a total of 207 countries participated in Olympics.
3. France, Italy, Switzerland, UK participated in all of the Olympics games.
4. Athletics, Cycling, Fencing, Gymnastics, Swimming were always present.
5. Aeronautics, Basque Pelota, Cricket and other 7 games had been removed from the Olympics.
6. A total of 34 games now played in Olympics and the least number of games played was just 8.
7. Oscar Gomer Swahn is oldest athlete (age: 64) ever to win a gold medal in shooting.
8. In 1896 Olympics there were no female athletes but it saw a steady increase over the years and eventually in 2016 Olympics female participants made a 45% of total athlete population.
9. Michael Fred Phelps, II won maximum Gold medals, 23 in Swimming
10. Heikki Ilmari Savolainen is higest medal scoring athlete (39) in Olympics.
11. USA won the most number of medals of all time, followed by Germany, France, UK and then Russia.
12. USA was the most active country scoring the most number of medals in almost every Olympic game since 1896.
13. Iceland, Malasiya, Philippines and 8 others country won Silver and Bronze, but never won any Gold medal. 
14. India won a total of 345 medals in Hockey.
15. India's performance in Hockey has declined over the years and they got only one 1 medal in 2008 Olympics.
